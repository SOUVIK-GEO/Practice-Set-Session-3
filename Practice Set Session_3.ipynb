# Create a 5X2 integer array and Prints its attributes

```python
import numpy as np

firstArray = np.array([[11, 34], [44, 50], [45, 89], [78, 89], [34, 78]]) 
print("Printing Array")
print(firstArray)

print("Printing numpy array Attributes")
print("1> Array Shape is: ", firstArray.shape)
print("2>. Array dimensions are ", firstArray.ndim)
print("3>. Length of each element of array in bytes is ", firstArray.itemsize)
```

# Create an array by adding the following two NumPy arrays. Next, modify the result array by calculating the square of each element

```python
import numpy as np
arrayOne = np.array([[5, 6, 9], [21, 18, 27]])
arrayTwo = np.array([[15, 33, 24], [4, 7, 1]])
print("Array One:")
print(arrayOne)
print("Array Two:")
print(arrayTwo)

sum_num = np.add(arrayOne, arrayTwo)
print("Array after Summation:")
print(sum_num)

sq_array = (sum_num)*(sum_num)
print("Square Array after Summation:")
print(sq_array)
```

# Find the length of each string element in the Numpy array

```python
import numpy as np
st_array = np.array(['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])

sizes = []
for i in st_array:
    sizes.append(len(i))

print(sizes)
```

# Find Mean ,median and mode of a List of Numpy Array

```python
import numpy as np
array = [78, 67, 87, 56, 23, 78, 98, 90, 455, 86, 23, 45, 87, 98, 90, 78]
mean = np.mean(array)
print(f"Mean: {mean:.2f}")
median = np.median(array)
print(f"Median: {median:.2f}")

from scipy import stats as st

print("Mode:", st.mode(array))
```

# Calculate the Euclidean distance using NumPy

```python
import numpy as np


point1 = np.array((2, 3))
point2 = np.array((5, 6))

dist = np.linalg.norm(point1 - point2)


print(dist)
```

# COVID-19 Data Visualization using matplotlib in Python

```python
import pandas as pd
df = pd.read_csv("C:/Users/Souvik Kundu/Desktop/5 Weeks Workshop Python July 2024/country_wise_latest.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = df['Confirmed']
y = df['Deaths']

plt.scatter(x, y, color = 'red')


z = np.polyfit(x, y, 1)
p = np.poly1d(z)


plt.plot(x, p(x))
```

```python
import matplotlib.pyplot as plt
x = df['WHO Region']
y = df['Deaths']

plt.bar(x, y, color = 'green')
plt.title('WHO Region-wise Number of Deaths due to Covid-19')
plt.xlabel('WHO Regions')
plt.ylabel('Number of Deaths')
plt.xticks(rotation = 90)

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(3)

plt.show()
```
